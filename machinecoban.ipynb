```python
# Cài 1 số gói hỗ trợ
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Đọc dữ liệu
data = pd.read_csv('aus.csv')
```

```python
data.head()
```

```python
X = data.iloc[:, :14]
y = data.Target
```

```python
X
```

```python
# Chia dữ liệu 2 tập train-test
# Từ tập train -> ước lượng w
# Từ tập test -> kiểm định w (hàm f) tìm được có tốt ko
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# ước lượng w nhờ X_train, y_train
X_train = X_train.values
X_train = X_train.transpose()
y_train = y_train.values
X_test = X_test.transpose()
X_train = np.concatenate((np.ones((1, X_train.shape[1])), X_train), axis = 0)
X_test = np.concatenate((np.ones((1, X_test.shape[1])), X_test), axis = 0)
```

```python
X_train
```

```python
X_test
```

```python
def sigmoid(s):
    return 1/(1 + np.exp(-s))

def logistic_sigmoid_regression(X, y, w_init, eta, tol = 1e-4, max_count = 500000):
    w = [w_init]    
    it = 0
    N = X.shape[1]
    d = X.shape[0]
    count = 0
    check_w_after = 20
    while count < max_count:
        # mix data 
        mix_id = np.random.permutation(N)
        for i in mix_id:
            xi = X[:, i].reshape(d, 1)
            yi = y[i]
            zi = sigmoid(np.dot(w[-1].T, xi))
            w_new = w[-1] + eta*(yi - zi)*xi
            count += 1
            # stopping criteria
            if count%check_w_after == 0:                
                if np.linalg.norm(w_new - w[-check_w_after]) < tol:
                    return w
            w.append(w_new)
    return w
```

```python
eta = .001 
d = X_train.shape[0]
w_init = np.random.randn(d, 1)

w = logistic_sigmoid_regression(X_train, y_train, w_init, eta)
```

```python
w[-1]
```

```python
# <0.5 -> 0;  >= 0.5 -> 1
def predict(f):
    f = f.reshape(-1)
    f = pd.DataFrame(f, columns = ['sigmoid'])
    #print(f)
    f['Target_pred'] = 0
    f.loc[f[f['sigmoid'] >= 0.5].index, 'Target_pred'] = 1
    return f
```

```python
# Dự đoán khách hàng nằm trong tập X_test
# Tìm độ chính mô hình logistic ??? w có tố k?
f = sigmoid(np.dot(w[-1].T, X_test))
target_pred = predict(f)
```

```python
target_pred
```

```python
# Đánh giá mô hình bằng độ chính xác
from sklearn.metrics import accuracy_score
accuracy_score(y_test.values, target_pred.Target_pred.values)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test.values, target_pred.Target_pred.values)
```
